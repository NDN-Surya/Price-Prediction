```python
# Define age bins and labels for the age groups
age_bins = [18, 25, 35, 45, 55, 70, float('inf')]
age_labels = ['18-25', '26-35', '36-45', '46-55', '56-70', '70+']

# Re-create 'age_group' with left-inclusive bins to ensure age 18 is included
df_new['age_group'] = pd.cut(df_new['age'], bins=age_bins, labels=age_labels, right=False)

# Remove the original 'age' column
# Drop the 'age' column
df_new = df_new.drop(columns=['age'])

# Display the updated DataFrame to verify
df_new.head()


# Display the updated DataFrame to verify
df_new.head()
```

```python
# Define the mappings for 'consume_frequency(weekly)' and 'awareness_of_other_brands'
consume_frequency_mapping = {
    "0-2 times": 1,
    "3-4 times": 2,
    "5-7 times": 3
}

awareness_mapping = {
    "0 to 1": 1,
    "2 to 4": 2,
    "above 4": 3
}

# Map the values in each column
df_new['frequency_score'] = df_new['consume_frequency(weekly)'].map(consume_frequency_mapping)
df_new['awareness_score'] = df_new['awareness_of_other_brands'].map(awareness_mapping)

# Calculate 'cf_ab_score' using the specified formula
df_new['cf_ab_score'] = ((df_new['frequency_score']) / ((df_new['awareness_score']) + (df_new['frequency_score']))).round(2)

# Drop the temporary columns if needed
df_new = df_new.drop(columns=['frequency_score', 'awareness_score'])


```

```python
# Define the mappings for 'zone' and 'income_levels'
zone_mapping = {
    "Urban": 3,
    "Metro": 4,
    "Rural": 1,
    "Semi-Urban": 2
}

income_levels_mapping = {
    "<10L": 1,
    "10L - 15L": 2,
    "16L - 25L": 3,
    "26L - 35L": 4,
    "> 35L": 5,
    "Not Reported": 0
}

# Map the values in each column
df_new['zone_score'] = df_new['zone'].map(zone_mapping)
df_new['income_score'] = df_new['income_levels'].map(income_levels_mapping)

# Calculate 'zas_score' as the product of zone_score and income_score
df_new['zas_score'] = df_new['zone_score'] * df_new['income_score']

# Drop the temporary columns if needed
df_new = df_new.drop(columns=['zone_score', 'income_score'])

# Display the updated DataFrame to verify
df_new[['zone', 'income_levels', 'zas_score']].head()
```

We assume age_group already exists in the DataFrame.

Retired individuals should not be in the 18-25 or 26-35 age groups.
The tilde ~ negates the condition, so we keep only the rows where the conditions are not true, i.e., valid records.
The cleaned DataFrame df_cleaned will no longer contain these outlier records.

```python
from sklearn.preprocessing import LabelEncoder

# Initialize LabelEncoder
le = LabelEncoder()

# List of columns to encode
columns_to_encode = ['age_group', 'income_levels', 'health_concerns', 'consume_frequency(weekly)', 'preferable_consumption_size']

# Apply Label Encoding to each column
for column in columns_to_encode:
    df_cleaned[column] = le.fit_transform(df_cleaned[column])

# Display the updated DataFrame to verify
df_cleaned.head()
```

```python
# List of columns to get unique values
columns_of_interest = [
    'gender', 'zone', 'occupation', 'current_brand', 'awareness_of_other_brands', 
    'reasons_for_choosing_brands', 'flavor_preference', 'purchase_channel', 
    'packaging_preference', 'typical_consumption_situations'
]

# Get unique values for each column
unique_values_dict = {column: df_cleaned[column].unique() for column in columns_of_interest}

# Display the unique values for each column
for column, unique_values in unique_values_dict.items():
    print(f"Unique values in {column}: {unique_values}")
```

```python
# List of columns to apply One-Hot Encoding
columns_to_encode = [
    'gender', 'zone', 'occupation', 'current_brand', 'awareness_of_other_brands', 
    'reasons_for_choosing_brands', 'flavor_preference', 'purchase_channel', 
    'packaging_preference', 'typical_consumption_situations'
]

# Apply One-Hot Encoding
df_clean_encoded = pd.get_dummies(df_cleaned, columns=columns_to_encode, drop_first=False)

# Display the updated DataFrame to verify
df_clean_encoded.head()
```

```python
# Remove the 'respondent_id' column
df_clean_encoded = df_clean_encoded.drop(columns=['respondent_id'])

# Display the updated DataFrame
df_clean_encoded.head()

from sklearn.preprocessing import LabelEncoder

# Initialize the LabelEncoder
le = LabelEncoder()

# Label encode the 'price_range' column
df_clean_encoded['price_range'] = le.fit_transform(df_clean_encoded['price_range'])

# Display the updated DataFrame
df_clean_encoded.head()
```

```python
# Assuming df_cleaned is your DataFrame after cleaning the data

# Prepare the feature matrix X and the target variable y
X = df_clean_encoded.drop(columns=['price_range'])  # Drop the respondent_id and target column
y = df_clean_encoded['price_range']  # Target variable (what you're predicting)

from sklearn.model_selection import train_test_split

# Split the dataset into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

# Check the sizes of the resulting datasets
print(f"Training set size: {X_train.shape}, Test set size: {X_test.shape}")
```

```python
# Import necessary libraries
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
import lightgbm as lgb
import xgboost as xgb
from sklearn.metrics import accuracy_score, classification_report

# Initialize the models
models = {
    "Gaussian Naive Bayes": GaussianNB(),
    "Logistic Regression": LogisticRegression(max_iter=1000),
    "SVM (Support Vector Machine)": SVC(),
    "Random Forest": RandomForestClassifier(random_state=42),
    "XGBoost": xgb.XGBClassifier(random_state=42)
}

# Train and evaluate each model
for model_name, model in models.items():
    # Fit the model
    model.fit(X_train, y_train)
    
    # Predict on the test set
    y_pred = model.predict(X_test)
    
    # Calculate accuracy
    accuracy = accuracy_score(y_test, y_pred)
    
    # Print model name and accuracy
    print(f"{model_name} Accuracy: {accuracy:.4f}")
    
    # Print Classification Report
    print(f"{model_name} Classification Report:")
    print(classification_report(y_test, y_pred))
    print("="*50)  # Separator between models
```
